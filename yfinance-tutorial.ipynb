```python
import yfinance as yf
```

```python
# Download data for a single ticker
data = yf.download("AAPL", start="2020-01-01", end="2021-01-01")
data.head(5)
```

```python
# Create a Ticker object for Apple
apple = yf.Ticker("AAPL")
```

```python
# Get stock info
apple.info
```

```python
# Get historical market data
hist = apple.history(period="1mo")
hist
```

```python
# Show dividends
apple.dividends
```

```python
# Show stock splits
apple.splits
```

```python
# Download data for multiple tickers
tickers = yf.download(["AAPL", "MSFT", "GOOGL"], start="2020-01-01", end="2021-01-01")
tickers.head
```

```python
# Access options data
options = apple.options
options
```

```python
# Get option chain for a specific expiration date
opt = apple.option_chain('2024-07-26')
print(opt.calls)
print(opt.puts)
```

```python
# Access quarterly financials
print(apple.quarterly_financials)
```

```python
# Access sustainability data
print(apple.sustainability)
```

```python
# Access analysts recommendations
print(apple.recommendations)
```

```python
import matplotlib.pyplot as plt

# Plot historical data
apple.history(period="5y")['Close'].plot(title="Apple Stock Price")
plt.show()
```
